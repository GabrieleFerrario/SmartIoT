```python
# link dataset: https://www.kaggle.com/muthuj7/weather-dataset
import pandas as pd
data = pd.read_csv('weatherHistory.csv')  
data.head()
```

```python
data.shape
```

```python
data["Summary"].value_counts()
```

```python
data["Wind Speed (m/s)"]=data.apply(lambda x: (x["Wind Speed (km/h)"]*1000)/3600, axis=1)
```

```python
data["Visibility (m)"]=data.apply(lambda x: (x["Visibility (km)"]*1000), axis=1)
```

```python
data= data[["Summary", "Temperature (C)", "Humidity", "Wind Speed (m/s)", "Pressure (millibars)"]]
data.head()
```

```python
data["Summary"].value_counts()
```

```python
temp = [['sun', 25.21, 0.51, 0.45, 1018], ['sun', 25.79, 0.49, 2.68, 1013], ['sun', 30.09, 0.39, 3, 1008], ['sun', 22.3,0.52,1.24,1011], ['sun', 34.8, 0.18,3.18 ,1015], ["sun", 24.56,0.6, 0.89,1017],["sun",35.96,0.18,3.17,1012], ["sun",26.8, 0.53, 3.05,1015],["sun",30.9, 0.40, 3.69,1010],["sun",31.63,0.25,4.64,1010], ["sun",27.12,0.46,0.89,1013],["sun",33.03,0.31,2.61,1008],["sun",34.83,0.22,1.37,1007],["sun",20.48999, 0.51, 2.68, 1001]]
```

```python
df2 = pd.DataFrame(temp, columns=['Summary', 'Temperature (C)', 'Humidity', 'Wind Speed (m/s)', 'Pressure (millibars)'])
```

```python
df2
```

```python
sun = data[(data["Summary"] == "Clear")].sample(n = 190)
sun=sun.append(df2)
sun = sun.append(data[(data["Summary"] == "Clear") & (data["Temperature (C)"] >25)].sample(n = 145))
sun["Summary"] = "sun"
print(sun.shape)
```

```python
cloud = data[(data["Summary"] == "Mostly Cloudy")]
cloud = cloud.sample(n = 700)
cloud["Summary"] = "cloudy"
cloud.shape
```

```python
rain = data[(data["Summary"] == "Light Rain") | (data["Summary"] == "Rain") | (data["Summary"] == "Drizzle")]
df_copy = data[(data["Summary"] == "Rain")].copy()
rain = rain.append(df_copy)
rain["Summary"] = "rain"
rain.shape
```

```python
final_df = (rain.append(sun)).append(cloud)
```

```python
final_df
```

```python
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X, y = oversample.fit_resample(final_df[["Temperature (C)","Humidity", "Wind Speed (m/s)", "Pressure (millibars)"]].values, final_df["Summary"].values)
```

```python
from collections import Counter

counter = Counter(y)
print(counter)
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0) # 70% training and 30% test
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_test, X_validation, y_test, y_validation = train_test_split(X_test, y_test, test_size=0.25,random_state=0) # 70% training and 30% test
```

```python
X_validation.shape
```

```python
X_test.shape
```

```python
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(C= 1000, gamma= 0.01, kernel= 'rbf')

#Train the model using the training sets
clf.fit(X_train, y_train)

```

```python
from sklearn import metrics
#Predict the response for test dataset
y_pred = clf.predict(X_test)
# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
clf.predict([[35.21, 0.71,0.45,1018]])# correct
```

```python
clf.predict([[17, 0.5, 5.68, 1011]])# wrong --> sun
```

```python
clf.predict([[21.73999, 0.94, 0.9, 1011]])# wrong --> rain
```

```python
from micromlgen import port

print(port(clf, classmap={
        0: 'cloudy',
        1: 'sun',
        2: 'rain'
    }))
```

```python
from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear', 'rbf', 'sigmoid']} 
grid = GridSearchCV(svm.SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_validation, y_validation)
```
